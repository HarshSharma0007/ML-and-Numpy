***Summerize Data ***

````python
# For each unique value in that column it count NO. or rows
Mainroad_price['price'].value_counts().head(5)
````

````python
# Number  distinct values in a column.
Mainroad_price['price'].nunique()
````

````python
# min value in that series or column 
Mainroad_price["price"].min()

# max value in that series or column
Mainroad_price["price"].max()

# average value in that series or column
Mainroad_price["price"].mean()

# median value in that series or column
Mainroad_price["price"].median()

# Standard deviation of each object or value 
Mainroad_price["price"].std()

# Varience of each object or value 
Mainroad_price["price"].var()

# Sum of each object or value 
Mainroad_price["price"].sum()

# It counts non null value
Mainroad_price["price"].count()

# Quantile of each object or value
# A quantile is a value that divides a probability distribution 
# or sorted data into equal parts.
Mainroad_price["price"].quantile() 

# Apply function to each object.
def double(integer):
    return 2*integer

Mainroad_price.apply(double)
````
